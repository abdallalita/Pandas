### Creating a dictionary

```python
wine={"white-wine":[3,4,5,6],"red-wine":[4,9,10,12]}
print(wine)
```

```python
#Converting to a dataframe

wine_drimes=pd.DataFrame(wine)
print(wine_drimes)
```

```python
#Extracting the indexes

wine_drimes.index
```

```python
#convert indexes (0 to 3) to (store1 to store4)

wine_drimes=pd.DataFrame(wine,index=["store1","store2","store3","store4"])
print(wine_drimes)
```

```python
#Extract the indexes

wine_drimes.index
```
